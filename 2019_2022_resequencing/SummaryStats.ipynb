```python
species=(sample_data['Species']!='Unknown') & (sample_data['Species']!='DLAB') & (sample_data['Species']!='DL') & (sample_data['Species']!='CNAT')
```

```python
#save as df and save target species into list
samples=sample_data[species]
species_list = samples["Species"].unique().tolist()
print(species_list)
```

```python
#drop any AS samples 
samples = samples[samples['NewTagNum'] != "AS"]
```

```python
# add up num of samples in each species 

# empty list to store data
summary_data=[]
# loop through each species for num of samples
for species in species_list:
    # create new df with only target species to reference
    df=samples[samples["Species"]==species]
    # for each health status, sum number of samples and output 
    #make list of health statuses
    statuses=df["Health_status"].unique().tolist()
    #loop thru each status and sum total number of samples 
    for x in statuses:
        num_rows = len(df[df["Health_status"] == x])
        # and by year 
        samps2019 = df[(df["Health_status"] == x) & (df["Year"] == '2019')]
        samps2022 = df[(df["Health_status"] == x) & (df["Year"] == '2022')]
        sum2019 = len(samps2019)
        sum2022 = len(samps2022)
        summary_data.append({'Species': species, 'Health_Status': x, 'Total_Samples': num_rows, 'Samples_2019': sum2019, 'Samples_2022': sum2022})
summary_df = pd.DataFrame(summary_data)
# Group by Species and sum the Total_Samples
total_samples_per_species = summary_df.groupby('Species')['Total_Samples'].sum().reset_index()
```

```python
health_status_order = ["Healthy", "Diseased_Tissue", "Diseased_Margin"]
# Convert Health_Status column to a categorical type with the specified order
summary_df['Health_Status'] = pd.Categorical(summary_df['Health_Status'], categories=health_status_order, ordered=True)
# Sort the DataFrame by Species and the ordered Health_Status
summary_df = summary_df.sort_values(by=['Species', 'Health_Status'])
```
