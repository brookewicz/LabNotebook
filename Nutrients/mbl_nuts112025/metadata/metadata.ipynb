```r
samples<-read.csv('mbl_nutrient_list.csv')
head(samples)
```

```r
samples<-samples %>%
    mutate(Date = as.Date(Date,  format = "%m/%d/%y")) %>%
    mutate(monthyear = format(Date, "%b %Y")) %>%
    select(-c(X,Pack_for_MBL))
```

```r
unique(samples$monthyear)
```

```r
# may and june 2025 are the same trip so change label 
samples[samples$monthyear=='May 2025',"monthyear"]<-"Jun 2025"
```
